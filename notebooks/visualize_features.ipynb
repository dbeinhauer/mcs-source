```python
import pickle
import matplotlib.pyplot as plt
```

```python
with open('../testing_dataset/pos_ori_phase_dictionary.pickle', 'rb') as f:
    feature_dict = pickle.load(f)
```

```python
for layer_name, layer_features in feature_dict.items():
    for feature_name in set(layer_features) - {'x', 'y'}:
        plt.figure(figsize=(5,5), constrained_layout=True)
        plt.scatter(layer_features['x'], layer_features['y'], c=layer_features[feature_name], cmap='hsv', s=5)
        if feature_name == 'ori':
            feature_name = 'orientation'
        plt.xlabel('X Position (mm)')
        plt.ylabel('Y Position (mm)')
        plt.title(f'{feature_name.capitalize()} Scatter Plot For Layer {layer_name}')
        plt.colorbar(label=f'{feature_name.capitalize()} (radians)', fraction=0.05, pad=0.05)
        plt.xlim(-2.5, 2.5)
        plt.ylim(-2.5, 2.5)
        plt.gca().set_aspect(1.0)
        plt.show()
```
