```python
class ModelVariants(Enum):
    SIMPLE_10 = "simple_10.0"
    DNN_10 = "dnn_10.0"
    SYN_ADAPT_10 = "syn_adapt_10.0"
    SIZE_0_3 = "model_0.3"
    SIZE_1 = "model_1.0"
    SIZE_10 = "syn_adapt_10.0"
    SIZE_1_3_VIS_33 = "model_1.3_visible_0.33"
    SIZE_2_VIS_50 = "model_2.0_visible_0.5"
    SIZE_11_VIS_90 = "model_11.0_visible_0.9"
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.lines as mlines
import numpy as np  # Needed for np.linspace

def plot_comparison_synchrony_curves_across_layers(df: pd.DataFrame, save_fig: str = ""):
    """
    Plot synchrony curves across layers (columns) for multiple models (rows),
    labeling only the leftmost plots of each row with A, B, C and adding a bottom legend.
    """
    ordered_layers = ["V1_Exc_L4", "V1_Exc_L23", "V1_Inh_L4", "V1_Inh_L23"]
    df["layer_name"] = pd.Categorical(df["layer_name"], categories=ordered_layers, ordered=True)
    
    pretty_layer_names = {
        "V1_Exc_L4": "L4 Excitatory ",
        "V1_Exc_L23": "L2/3 Excitatory ",
        "V1_Inh_L4": "L4 Inhibitory ",
        "V1_Inh_L23": "L2/3 Inhibitory ",
    }

    ordered_models = [
        ModelVariants.SIMPLE_10.value,
        ModelVariants.DNN_10.value,
        ModelVariants.SYN_ADAPT_10.value,
        # ModelVariants.SIZE_1.value,
        # ModelVariants.SIZE_2_VIS_50.value,
        ModelVariants.SIZE_0_3.value,
    ]
    row_labels = {
        ModelVariants.SIMPLE_10.value: 'Simple (10,800 neurons)', 
        ModelVariants.DNN_10.value: 'DNN (10,800 neurons)', 
        ModelVariants.SYN_ADAPT_10.value: 'RNN (10,800 neurons)', 
        ModelVariants.SIZE_1.value: 'RNN (1,080 neurons)',
        ModelVariants.SIZE_2_VIS_50.value: 'RNN (1,080 neurons visible \n+ 1,080 neurons invisible)',
        ModelVariants.SIZE_0_3.value: 'RNN (324 neurons)',
    }
    df["model_variant"] = pd.Categorical(df["model_variant"], categories=ordered_models, ordered=True)
    df.columns.name = None
    df["time_ms"] = df["time"] * 20

    g = sns.relplot(
        data=df,
        x="time_ms",
        y="synchrony",
        kind="line",
        hue="variant_type",
        style="variant_type",
        row="model_variant",
        col="layer_name",
        facet_kws={"sharey": 'col', "sharex": True},
        height=4,
        aspect=1.65,
        linewidth=3,
        errorbar=None,
        palette="colorblind"
    )

    # Add vertical stimulus change line
    def add_vline(data, color, **kwargs):
        plt.axvline(
            nn_model.globals.IMAGE_DURATION,
            color="gray",
            linestyle="--",
            linewidth=2.2,
        )
    g.map_dataframe(add_vline)

    # Add grid lines
    g.map(lambda *args, **kwargs: plt.grid(True, linestyle=":", linewidth=0.4, alpha=0.6))

    # Axis labels
    g.set_axis_labels("Time (ms)", "Ratio of Spiking Neurons", size=20)
    g.set_titles("")  # Remove col/row template titles

    # Annotate each row
    for i, ax_row in enumerate(g.axes):
        ax = ax_row[0]
        ax.annotate(
            row_labels[ordered_models[i]],
            xy=(0, 1.10),
            xycoords='axes fraction',
            xytext=(-0.15, 1.16),
            textcoords='axes fraction',
            fontsize=26,
            fontweight='bold',
            va='bottom',
            ha='left',
            annotation_clip=False
        )

    # Y-axis: 4 ticks per subplot
    num_ticks = 4
    for ax in g.axes.flat:
        ymin, ymax = ax.get_ylim()
        fixed_yticks = np.linspace(ymin, ymax, num_ticks)
        ax.set_yticks(fixed_yticks)
        ax.set_yticklabels([f"{y:.4f}" for y in fixed_yticks], fontsize=16)
        ax.tick_params(axis='both', labelsize=16)

    # Column titles
    for i, ax in enumerate(g.axes[0]):
        ax.set_title(pretty_layer_names[ordered_layers[i]], fontsize=28)

    g.figure.tight_layout(rect=[0.05, 0, 1, 1])  # Leave space at bottom for legend

    # Custom legend at bottom, line by line
    stimulus_line = mlines.Line2D([], [], color="gray", linestyle="--", linewidth=2, label="Stimulus change")
    separator = mlines.Line2D([], [], color="black", linestyle="-", linewidth=0, label="────────────")
    handles, labels = g.axes.flat[0].get_legend_handles_labels()
    g._legend.remove()

    g.figure.legend(
        handles + [separator, stimulus_line],
        ["Predictions", "Target", "────────────", "Stimulus change"],
        loc="lower center",
        fontsize=20,
        title="Variant Type + Marker",
        title_fontsize=21,
        frameon=True,
        ncol=4,
        bbox_to_anchor=(0.5, -0.06)
    )

    if save_fig:
        g.figure.savefig(save_fig, format="pdf", bbox_inches="tight")
    else:
        plt.show()
```

```python
PLOTING_DESTINATION = "/home/david/source/diplomka/thesis_plots/"
EVALUATION_RESULTS_BASE = "/analysis_results"
CONFERENCE_BASE = "bernstein_conference/"
analysis_paths = {
    EvaluationProcessorChoices.FULL_DATASET_ANALYSIS: f"{nn_model.globals.PROJECT_ROOT}{EVALUATION_RESULTS_BASE}/{EvaluationProcessorChoices.FULL_DATASET_ANALYSIS.value}/",
    EvaluationProcessorChoices.SUBSET_DATASET_ANALYSIS: f"{nn_model.globals.PROJECT_ROOT}{EVALUATION_RESULTS_BASE}/{EvaluationProcessorChoices.SUBSET_DATASET_ANALYSIS.value}/",
    EvaluationProcessorChoices.WANDB_ANALYSIS: f"{nn_model.globals.PROJECT_ROOT}{EVALUATION_RESULTS_BASE}/{EvaluationProcessorChoices.WANDB_ANALYSIS.value}/results.pkl",
    EvaluationProcessorChoices.PREDICTION_ANALYSIS: f"{nn_model.globals.PROJECT_ROOT}{EVALUATION_RESULTS_BASE}/{EvaluationProcessorChoices.PREDICTION_ANALYSIS.value}/",
}
analysis_paths[EvaluationProcessorChoices.PREDICTION_ANALYSIS] += CONFERENCE_BASE
# result_analyzer = ResultAnalyzer(analysis_paths)

all_model_variants = [
    ModelVariants.SIMPLE_10,
    ModelVariants.DNN_10,
    ModelVariants.SYN_ADAPT_10,
    ModelVariants.SIZE_0_3,
    # ModelVariants.SIZE_1,
    # ModelVariants.SIZE_2_VIS_50,
]

all_model_dfs = []

for model_variant in all_model_variants:
    result_analyzer = ResultAnalyzer(
        analysis_paths, load_custom_variant=model_variant.value,
    )

    plot_data = result_analyzer.prepare_dataframe_for_plot(
        PlottingVariants.SEPARATE_TEMPORAL_BEHAVIOR_TARGET_PREDICTION,
        synchrony_curve_kwargs={
            "model_variants": [model_variant.value],
        },
    )
    if model_variant in [ModelVariants.SIZE_1_3_VIS_33, ModelVariants.SIZE_2_VIS_50, ModelVariants.SIZE_11_VIS_90]:
        plot_data = plot_data[plot_data['visibility_variant'] == VisibilityVariants.VISIBLE_NEURONS]

    plot_data['model_variant'] = model_variant.value
    diminisher = 1
    if model_variant == ModelVariants.DNN_10:
        diminisher = 7
    elif model_variant == ModelVariants.SYN_ADAPT_10:
        diminisher = 3
    elif model_variant == ModelVariants.SIMPLE_10:
        diminisher = 1 
    elif model_variant in [ModelVariants.SIZE_1, ModelVariants.SIZE_2_VIS_50]:
        diminisher = 0.1
    elif model_variant == ModelVariants.SIZE_0_3:
        diminisher = 0.03
    plot_data['synchrony'] /= diminisher
    plot_data['synchrony'] *= 20
    
    # plot_data["model_variant"] = model_name_pretty  # Add column for comparison
    all_model_dfs.append(plot_data)

combined_df = pd.concat(all_model_dfs, ignore_index=True)
plot_comparison_synchrony_curves_across_layers(combined_df, save_fig="poster/combined_comparison_plot.pdf")
```
